```python
from firebase_admin import credentials, firestore
from datetime import datetime
import firebase_admin
import logging
import json
import os
```

```python
recrawled_telegram_data_path = "../data/re/recrawled_telegram_data.json"

with open(recrawled_telegram_data_path, "r") as f:
    recrawled_telegram_data = json.load(f)
```

```python
# 파베 로그인
firebase_key_path = '../../data/firebase-key.json'

if not firebase_admin._apps:
    cred = credentials.Certificate(firebase_key_path)
    firebase_admin.initialize_app(cred)
db = firestore.client()
```

```python
# 로그 디렉토리 생성
log_dir = "../logs"
os.makedirs(log_dir, exist_ok=True)

# 로깅 설정
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler(f"{log_dir}/firebase_operations.log", encoding='utf-8'),
        logging.StreamHandler()  # 콘솔에도 출력
    ]
)

logger = logging.getLogger("firebase_operations")
```

```python
category1 = "try_sector_detail_page"
category2 = "detail_dates"
```

```python
# 섹터 데이터 저장
def save_sector_data(sector, date, data):
    if db.collection(category1).document(sector).collection(category2).document(date).get().exists:
        pass
    else:
        try:
            db.collection(category1).document(sector).collection(category2).document(date).set(data)
            logger.info(f"{sector} 데이터 저장 완료")
            return f"{sector} 데이터 저장 완료"
        except Exception as e:
            error_msg = f"{sector} 데이터 저장 중 오류 발생: {e}"
            logger.error(error_msg)
            return error_msg
```

```python
import random

for sector in recrawled_telegram_data.keys():
    for date in recrawled_telegram_data[sector].keys():
        for channel in recrawled_telegram_data[sector][date].keys():
            if "posts" in recrawled_telegram_data[sector][date][channel]:
                recrawled_telegram_data[sector][date][channel]["score"] = random.randint(0, 100)

        data_to_save = recrawled_telegram_data[sector][date]
        save_sector_data(sector, date, data_to_save)


```
